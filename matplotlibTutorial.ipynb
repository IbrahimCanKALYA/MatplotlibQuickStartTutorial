# *Çoklu Şekil Ve Eksenlerde Çalışma

```python
def func(x):
    return x*x
x=np.arange(0,5)
y=np.arange(6,11)

plt.plot(x,func(y),'--',label='numbers')#Bu şekilde plot un içine parametre olarak fonksiyon da verebilirsiniz...
#label parametresi ile çizgeye isim verebiliriz...
plt.legend()#Bu methodla birden fazla çizgenin karışmamasını sağlayacak aynı haritalardaki lejant gibi bir legend oluşturabiliriz...
plt.show()
```

```python
plt.figure(1)#Bu methodla tablo sayısını belirleyebiriz...
plt.subplot(211)#Bu methodla da tabloların içindeki altçizgeleri belirleyebiliriz..
plt.plot([1,2,3,4],[8,7,6,5],'-')
plt.subplot(211)
plt.plot([1,2,3,4],[5,6,7,8],'r-')
plt.figure(2)
plt.subplot(211)
plt.plot([1,2,3,4],[8,8,6,5],'y-')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D #3D veri görselleştirme aracını import edelim...
#Grafiklerin görükmesi için pylab aktifleştirilir...
%pylab inline     
```

# 3D scatter Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')#3D koordinat sistemi ayarlıyoruz...

z = np.linspace(0, 1, 100)#Verilerlimizi ayarlıyoruz...
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

c = x + y
#2 veya 3 boyutlu uzayda, bagimsiz degi$kenler arasi ili$kiyi gosteren istatistik grafigine scatter diagram da denir...
ax.scatter(x, y, z, c=c)#İlk 3 parametre dizleri son parametre ise renk dizisini belirtir...
```

# 3D Line Plot;

```python
#Çizgi grafik de aynı scatter gibi aynı parametrelere sahiptir o yüzden yukarıda çidiğimiz scatter grafiği çizgi graph a çevirelim...
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot(x,y,z,'-r')
```

# Surface Plot

 

```python
#Eksenlerin bağlantılı oduğu yüzey grafikleridir.
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))#outer methodu iki vektörün dış çarpımını hesaplar...
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)
#cmap:renk dizisi  rstride:satır basamağı  cstride:kolon basamağı  linewitdh:çizgi kalınlığı
```

# Wire-frame plot

```python
#Wire-frame ler birbirine yakın noktalar arasına çizgi çekerek grafik çiziyor.

u = np.linspace(0, np.pi, 30)
v = np.linspace(0, 2 * np.pi, 30)

x = np.outer(np.sin(u), np.sin(v))
y = np.outer(np.sin(u), np.cos(v))
z = np.outer(np.cos(u), np.ones_like(v))


fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z)#Bu şekilde yukarıda hazırladığımız veri setlerini vererek bir küre çizeceğiz...
```
